```python
#rescaling the data
from sklearn.preprocessing import StandardScaler
```

```python
se = StandardScaler()
```

```python
y = df_copy['Outcome']
```

```python
X = df_copy.drop(labels = 'Outcome', axis = 1)
```

```python
X
```

```python
se_data = se.fit_transform(X)
```

```python
X_col = X.columns
```

```python
X = pd.DataFrame(se_data, columns=X_col)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 43, stratify = y)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
test_score = []
train_score = []
```

```python
for i in range(1,16):
    knn  = KNeighborsClassifier(i)
    knn.fit(X_train, y_train)
    
    train_score.append(knn.score(X_train, y_train))
    test_score.append(knn.score(X_test, y_test))
```

```python
plt.figure(figsize=(8,6))
plt.plot(range(1,16), train_score, marker = '*', alpha = 0.3)
plt.plot(range(1,16), test_score, marker = 'o', linestyle = '--')
plt.legend(labels=["Train_Score", "Test_Score"], loc = 0)
```

```python
knn = KNeighborsClassifier(14)
```

```python
knn.fit(X_train, y_train)
knn.score(X_test, y_test)*100
```

```python
knn.score(X_train, y_train)*100
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.metrics import classification_report
```

```python
confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot = True, fmt=  'd')
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import roc_curve
```

```python
y_pred_prob = knn.predict_proba(X_test)[:,1]
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
```

```python
plt.plot(fpr, tpr)
plt.plot([0,1], [0,1], linestyle = '--', color = 'k')
```
